# Assigning Meaningful Column Names

The dataset was imported without descriptive column names. Assigning meaningful column names improves readability and makes subsequent analysis easier to understand and maintain.

```python
import pandas as pd

# Define column names from UCI dataset description
columns = [
    "symboling", "normalized_losses", "make", "fuel_type", "aspiration",
    "num_doors", "body_style", "drive_wheels", "engine_location",
    "wheel_base", "length", "width", "height", "curb_weight",
    "engine_type", "num_cylinders", "engine_size", "fuel_system",
    "bore", "stroke", "compression_ratio", "horsepower",
    "peak_rpm", "city_mpg", "highway_mpg", "price"
]

# Read the data
allo = pd.read_csv(url, names=columns)

# Check first rows
allo
```

# Handling Missing Values

In this dataset, missing values are represented by question marks (?). Since Pandas does not automatically recognize these as missing values, they must be converted to NaN (Not a Number) so they can be properly identified and treated during the cleaning process.

```python
import numpy as np

# replace "?" to NaN
allo.replace("?",np.nan,inplace=True)
allo
```

# Identifying Missing Data

After converting invalid entries to NaN values, we assess the dataset to determine which columns contain missing values and the extent of missingness in each feature.

```python
allo.isnull().sum()
```

```python
allo['normalized_losses']
```

# Imputing Missing Values in Normalized Losses

The normalized_losses column contains missing numerical values. To preserve the dataset and avoid losing observations, missing entries are replaced using the mean value of the column. Mean imputation is commonly used when dealing with continuous numerical variables.

```python
avg_normal_loss=allo['normalized_losses'].astype('float').mean(axis=0)
avg_normal_loss
```

```python
allo['normalized_losses'].replace(np.nan,avg_normal_loss, inplace=True)
allo.head()
```

```python
allo.isnull().sum()
```

# Handling Missing Values in Categorical Data

The num_doors column is a categorical feature. Since calculating an average is not meaningful for categorical variables, missing values are replaced using the mode, which represents the most frequently occurring category in the dataset.

```python
allo['num_doors']
```

```python
allo['num_doors'].value_counts()
```

```python
allo['num_doors'] = allo['num_doors'].fillna('four')
```

```python
allo.isnull().sum()
```

# Imputing Missing Values in Bore

The bore feature contains a small number of missing numerical values. These values are replaced using the column mean to maintain consistency while minimizing the impact on the overall distribution of the data.

```python
allo['bore']
```

```python
avg_bore =allo['bore'].astype('float').mean(axis=0).round(2)
avg_bore
```

```python
allo['bore'].replace(np.nan,avg_bore,inplace=True)
```

```python
allo.isnull().sum()
```

## Imputing Missing Values in Stroke

To ensure data completeness, missing values in the stroke column are replaced with the average stroke value calculated from the available observations.

```python
allo['stroke']
```

```python
avg_stroke =allo['stroke'].astype("float").mean(axis=0).round(2)
avg_stroke
```

```python
allo['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
allo.isnull().sum()
```

# Imputing Missing Values in Horsepower

Horsepower is an important numerical feature that influences vehicle performance. Missing values are replaced with the mean horsepower value to maintain the integrity of the dataset.

```python
allo["horsepower"]
```

```python
avg_horse =allo["horsepower"].astype("float").mean(axis=0).round(2)
avg_horse
```

```python
allo["horsepower"].replace(np.nan,avg_horse,inplace=True)
```

```python
allo.isnull().sum()
```

# Imputing Missing Values in Peak RPM

Missing values in the peak_rpm column are replaced using the column average. This approach preserves all observations while maintaining a reasonable representation of the feature.

```python
allo["peak_rpm"]
```

```python
avg_peak =allo["peak_rpm"].astype("float").mean(axis=0).round(2)
avg_peak
```

```python
allo["peak_rpm"].replace(np.nan,avg_peak,inplace=True)
```

```python
allo.isnull().sum()
```

# Removing Records with Missing Prices

Price is the target variable and one of the most important features in the dataset. Records with missing prices are removed because imputing the target variable could introduce bias and reduce the reliability of future analyses.

```python
allo.dropna(subset=["price"],inplace=True)
```
